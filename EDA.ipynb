```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8')
sns.set_palette('coolwarm')
```

```python
#Cleaning
df.drop(['license', 'house_rules', 'country code', 'Construction year', 'last review',
        'reviews per month', 'review rate number', 'availability 365'],axis=1,inplace=True)
df
```

```python
#Cleaning
df['price']=df['price'].replace('[\$,]', '',regex=True).astype(float)
df['service fee']=df['service fee'].replace('[\$,]', '',regex=True).astype(float)
```

```python
#Handling missing values
for col in df.select_dtypes(include=['float64','int64']).columns:
    df[col].fillna(df[col].mean(),inplace=True)
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0],inplace=True)
```

```python
print(df.isnull().sum())
print(df.dtypes)
```

# Price Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'],bins=50,kde=True,color='teal')
plt.title('Distribution of Airbnb Prices')
plt.xlabel('Price ($)')
plt.ylabel('Count')
plt.show()
```

# Price by Room type

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x='room type',y='price',palette='coolwarm')
plt.title('Price Rnge by Room type')
plt.xlabel('Room type')
plt.ylabel('Price ($)')
plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(8,5))
corr = df.drop(columns=['id', 'host id', 'lat', 'long'], errors='ignore').corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

# Price vs No. of reviews

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='number of reviews',y='price',alpha=0.6,color='skyblue')
plt.title('Price vs No. of reviews')
plt.xlabel('No. of reviews')
plt.ylabel('Price ($)')
plt.show()
```
